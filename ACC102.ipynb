```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings

warnings.filterwarnings('ignore')

# 设置字体（使用系统默认字体）
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

def create_in_memory_data():
    """在内存中创建示例财务数据"""
    data = {
        'Date': ['2023-12-31', '2022-12-31', '2021-12-31', '2020-12-31', '2019-12-31'],
        'Total Revenue': [383332000000, 394328000000, 365817000000, 274515000000, 260174000000],
        'Net Income': [96995000000, 99803000000, 94680000000, 57411000000, 55256000000],
        'Operating Income': [110512000000, 118234000000, 108949000000, 66288000000, 63930000000],
        'Total Assets': [352755000000, 352755000000, 351002000000, 323888000000, 338516000000],
        'Total Liab': [290437000000, 283221000000, 287912000000, 258549000000, 248028000000],
        'Total Current Assets': [135405000000, 135405000000, 134836000000, 143713000000, 162819000000],
        'Total Current Liabilities': [109331000000, 109331000000, 125481000000, 105392000000, 105718000000],
        'Total Stockholder Equity': [62318000000, 69534000000, 63090000000, 65339000000, 90488000000]
    }
    
    df = pd.DataFrame(data)
    return df

def load_financial_data():
    """加载财务数据（在内存中）"""
    data = create_in_memory_data()
    print("数据创建成功！")
    print("数据形状：", data.shape)
    print("\n数据前5行：")
    print(data.head())
    return data

def clean_financial_data(df):
    """清洗财务数据"""
    # 转换日期格式
    df['Date'] = pd.to_datetime(df['Date'])
    
    # 删除全空行
    df = df.dropna(how='all')
    
    # 填充缺失值
    df = df.fillna(method='ffill').fillna(method='bfill')
    
    return df

def calculate_financial_metrics(df):
    """计算财务指标"""
    # 净利润率
    df['NetProfitMargin'] = df['Net Income'] / df['Total Revenue']
    
    # 资产负债率
    df['DebtRatio'] = df['Total Liab'] / df['Total Assets']
    
    # 净资产收益率（ROE）
    df['ROE'] = df['Net Income'] / df['Total Stockholder Equity']
    
    # 流动比率
    df['CurrentRatio'] = df['Total Current Assets'] / df['Total Current Liabilities']
    
    # 营业利润率
    df['OperatingMargin'] = df['Operating Income'] / df['Total Revenue']
    
    return df

def visualize_financial_metrics(df):
    """可视化财务指标"""
    # 创建图表
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    
    # 净利润率
    axes[0, 0].plot(df['Date'], df['NetProfitMargin'])
    axes[0, 0].set_title('净利润率趋势')
    axes[0, 0].set_ylabel('净利润率')
    
    # 资产负债率
    axes[0, 1].plot(df['Date'], df['DebtRatio'])
    axes[0, 1].set_title('资产负债率趋势')
    axes[0, 1].set_ylabel('资产负债率')
    
    # ROE
    axes[1, 0].plot(df['Date'], df['ROE'])
    axes[1, 0].set_title('净资产收益率趋势')
    axes[1, 0].set_ylabel('ROE')
    
    # 流动比率
    axes[1, 1].plot(df['Date'], df['CurrentRatio'])
    axes[1, 1].set_title('流动比率趋势')
    axes[1, 1].set_ylabel('流动比率')
    
    plt.tight_layout()
    plt.show()

def display_html_report(df):
    """显示HTML报告内容（不保存到文件）"""
    html_content = f"""
    <!DOCTYPE html>
    <html lang="zh-CN">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>苹果公司财务分析报告</title>
        <style>
            body {{
                font-family: Arial, sans-serif;
                margin: 0;
                padding: 20px;
                background-color: #f5f5f5;
            }}
            .container {{
                max-width: 1200px;
                margin: 0 auto;
                background-color: white;
                padding: 20px;
                border-radius: 8px;
                box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
            }}
            h1 {{
                color: #333;
                text-align: center;
            }}
            img {{
                display: block;
                margin: 0 auto;
                max-width: 100%;
                height: auto;
                margin-bottom: 20px;
            }}
            table {{
                width: 100%;
                border-collapse: collapse;
                margin-top: 20px;
            }}
            th, td {{
                border: 1px solid #ddd;
                padding: 8px;
                text-align: left;
            }}
            th {{
                background-color: #f2f2f2;
            }}
        </style>
    </head>
    <body>
        <div class="container">
            <h1>苹果公司财务分析报告</h1>
            <p>报告生成时间: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}</p>
            
            <h2>财务指标趋势图</h2>
            <img src="financial_metrics.png" alt="财务指标趋势图">
            
            <h2>财务指标数据</h2>
            {df.to_html(classes='data', index=False)}
        </div>
    </body>
    </html>
    """
    
    print("\n=================== 财务分析报告 ===================")
    print(html_content)

if __name__ == "__main__":
    # 加载数据
    financial_data = load_financial_data()
    
    if financial_data is not None:
        # 清洗数据
        cleaned_data = clean_financial_data(financial_data)
        
        # 计算财务指标
        metrics_data = calculate_financial_metrics(cleaned_data)
        
        # 可视化财务指标
        visualize_financial_metrics(metrics_data)
        
        # 显示图表信息（不保存到文件）
        print("\n=================== 图表信息 ===================")
        print("图表已显示，但由于文件系统是只读的，无法保存")
        
        # 显示分析结果
        print("\n=================== 财务指标摘要 ===================")
        print("平均净利润率: {:.2%}".format(metrics_data['NetProfitMargin'].mean()))
        print("平均资产负债率: {:.2%}".format(metrics_data['DebtRatio'].mean()))
        print("平均ROE: {:.2%}".format(metrics_data['ROE'].mean()))
        print("平均流动比率: {:.2f}".format(metrics_data['CurrentRatio'].mean()))
        
        # 显示报告内容（不保存到文件）
        print("\n=================== 报告信息 ===================")
        print("报告内容将直接显示在下方（不保存到文件）")
        display_html_report(metrics_data)
```
